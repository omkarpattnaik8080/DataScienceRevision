```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.ylabel('sin(x)')
```

```python
plt.title("this is my title")
```

```python
x=np.linspace(0,10,200)
y=np.sin(x)
plt.plot(x,y,'--')
plt.xlabel('x')
plt.ylabel('y')
plt.title("this is my title")
```

```python
x=np.linspace(0,10,200)
plt.figure(figsize=(6,2))
y=np.sin(x)
plt.plot(x,y,'--')
plt.xlabel('x')
plt.ylabel('y')
plt.title("this is my title")
```

```python
x=np.linspace(0,10,200)
plt.figure(figsize=(6,2))
y=np.sin(x)
plt.plot(x,y,'--')
plt.xlabel('x')
plt.ylabel('y')
plt.title("this is my title")
plt.grid()
```

```python
x=np.random.randint(100)
y=np.random.randint(100)
colours=np.random.rand(50)
size=1000*np.random.rand(1) # change size to a single value
plt.scatter(x,y, c ="b",s= size,alpha=.1,marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title("this is my title")
```
