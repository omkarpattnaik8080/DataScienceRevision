```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
```

```python
#interpolate the data using Linear interpolar
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#if we want to add datapoints we will use linear interpolation
x_new=np.linspace(1,5,10)
y_inter=np.interp(x_new,x,y)
print(y_inter)
```

```python
plt.scatter(x_new,y_inter)
```

```python
#cubic interpolation with scipy
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
```

```python
from scipy.interpolate import interp1d
```

```python
interp1d(x,y,kind='cubic')
```

```python
#interpolate the data
x_new=np.linspace(1,5,10)
y_inter=np.interp(x_new,x,y)
print(y_inter)
```

```python
plt.scatter(x_new,y_inter)
```

```python
#polynomial interpolation
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
```

```python
#interpolate the data using polynomial interpolation
p=np.polyfit(x,y,2)
print(p)
```

```python
x_new=np.linspace(1,5,10)
y_inter=np.polyval(p,x_new)
print(y_inter)
```

```python
plt.scatter(x_new,y_inter)
```
