```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris)
```

```python
#acts as a histogram
sns.displot(iris['sepal_length'])
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts() # Changed values_counts to value_counts
```

```python
#to know who are smoker and who are non smoker
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='size',data=tips,style='size')
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(iris)
```
